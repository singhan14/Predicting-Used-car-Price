# Observation 

A Q-Q (quantile-quantile) plot is a graphical tool used to assess whether a data set follows a specified theoretical distribution, typically a normal distribution.

- The x-axis represents the theoretical quantiles (expected values if the data were perfectly normal).  
- The y-axis represents the observed quantiles from the data.  

**If the data closely follows the normal distribution, the points should lie along a 45-degree reference line. However, if the points deviate significantly from this line, it suggests the data may not be normally distributed.**

In this case, since the points are diverging from the 45-degree line, this could indicate one or more of the following issues with normality:  

1. **Skewness**: The distribution might be asymmetrical, with a longer tail on one side.  
2. **Kurtosis**: The data may have heavier or lighter tails than expected under normality.  
3. **Outliers**: Extreme values could be pulling the points away from the line.  
4. **Multimodality**: The data might contain multiple peaks, deviating from the single peak of a normal distribution.  

These deviations may necessitate statistical tests (e.g., Shapiro-Wilk, Anderson-Darling) to confirm the lack of normality or further data transformations to meet the assumptions of normality if required by your analysis.

# BOOTSTRAPING THE MODEL

In my analysis, I observed that the residuals from my model did not follow a normal distribution, as indicated by the Q-Q plot. To address this issue, I implemented a series of diagnostic and corrective steps:

1. **Normality Diagnostics**:
   - Constructed a Q-Q plot to assess the normality of residuals, which revealed significant deviations from normality.
   - Conducted the Shapiro-Wilk test for normality, confirming the non-normality of residuals.

2. **Data Transformations**:
   - Applied a Box-Cox transformation to the dependent variable to stabilize variance and improve normality.
   - Standardized the target variable using a StandardScaler, which rescaled the data but still failed to achieve normality.

3. **Studentization**:
   - Transformed the residuals into studentized residuals. Despite this adjustment, the normality assumption was still not satisfied.

4. **Bootstrap Resampling**:
   - As a final approach, I employed bootstrap resampling techniques to address the non-normality. Bootstrapping allowed me to generate robust estimates of the model parameters and confidence intervals without relying on the normality assumption.

## Bootstraping Model 1

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

response_variable = 'selling_price'  

predictors = car.columns.difference([response_variable]) 

def bootstrap_model(car, response_variable, predictors, num_samples):
    """Bootstrap the regression model and return the coefficients."""
    bootstrapped_coefs = []
    n = len(car)

    for _ in range(num_samples):
        sample = car.sample(n=n, replace=True)
        X = sample[predictors]
        X = sm.add_constant(X)  
        y = sample[response_variable]

        model = sm.OLS(y, X).fit()
        bootstrapped_coefs.append(model.params)

    return pd.DataFrame(bootstrapped_coefs)

num_samples = 1000  

bootstrapped_results = bootstrap_model(car, response_variable,
predictors, num_samples)

mean_coefs = bootstrapped_results.mean()
lower_bound = bootstrapped_results.quantile(0.025)
upper_bound = bootstrapped_results.quantile(0.975)

print("Bootstrapped Coefficients with 95% Confidence Intervals:")
results_df = pd.DataFrame({
    'Mean Coefficient': mean_coefs,
    'Lower Bound': lower_bound,
    'Upper Bound': upper_bound
})
print(results_df)

plt.figure(figsize=(10, 5))
plt.hist(bootstrapped_results.iloc[:, 1], bins=30, edgecolor='k',
alpha=0.7)  
plt.axvline(mean_coefs[1], color='red', linestyle='dashed',
linewidth=2, label='Mean Coefficient')
plt.axvline(lower_bound[1], color='blue', linestyle='dashed',
linewidth=2, label='95% CI Lower Bound')
plt.axvline(upper_bound[1], color='blue', linestyle='dashed',
linewidth=2, label='95% CI Upper Bound')
plt.title('Distribution of Bootstrapped Coefficients for Predictor')
plt.xlabel('Coefficient Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

## Mallow CP

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from itertools import combinations

X = car.drop('selling_price', axis=1)  
y = df['selling_price']  

full_model = LinearRegression().fit(X, y)
full_rss = np.sum((y - full_model.predict(X))**2) 
full_p = len(X.columns)  

def calculate_mallows_cp(X, y, full_rss, full_p):
    """
    Calculate Mallows' Cp for all subsets of predictors.

    Args:
        X (pd.DataFrame): Predictor variables.
        y (pd.Series): Response variable.
        full_rss (float): RSS of the full model.
        full_p (int): Number of predictors in the full model.

    Returns:
        pd.DataFrame: Table of subsets with their Mallows' Cp.
    """
    n = len(y)
    predictors = X.columns
    results = []

    for k in range(1, len(predictors) + 1):
        for subset in combinations(predictors, k):
            subset_X = X[list(subset)]
            model = LinearRegression().fit(subset_X, y)
            rss = np.sum((y - model.predict(subset_X))**2)
            p = len(subset)  

            cp = rss / (full_rss / (n - full_p)) - (n - 2 * p)
            results.append({
                'Subset': subset,
                'RSS': rss,
                'Cp': cp
            })

    results_df = pd.DataFrame(results)
    return results_df

mallows_cp_results = calculate_mallows_cp(X, y, full_rss, full_p)

print(mallows_cp_results.sort_values(by='Cp'))
mallows_cp_results['Subset'][125]
```

```python
X = car2[['engine', 'km_driven', 'max_cost_price', 'max_power', 'mileage', 'vehicle_age']]
y = car2['selling_price']

X = np.c_[np.ones(X.shape[0]), X]  

X_transpose = X.T
X_transpose_X = X_transpose @ X

X_transpose_X_inv = np.linalg.inv(X_transpose_X)

hat_matrix = X @ X_transpose_X_inv @ X_transpose

leverage_values = np.diagonal(hat_matrix)

car2['leverage'] = leverage_values

print(leverage_values)
print(f"Number of leverage points detected: {sum(leverage_values > 2 * (X.shape[1] / len(X)))}")
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

unrestricted_model = ols("selling_price ~ engine + km_driven + max_cost_price + max_power + mileage + vehicle_age", data=car2).fit()

null_model = ols("selling_price ~ 1", data=car2).fit()
anova_results = sm.stats.anova_lm(null_model, unrestricted_model)

print(anova_results)
```

# Observation

The ANOVA results compare a null model (with only an intercept) to an unrestricted model (with predictors like `engine`, `km_driven`, etc.).

1. **Degrees of Freedom for Residuals (df_resid):**
   - Null model: 15,410 (total observations minus 1).
   - Unrestricted model: 15,404 (total observations minus the number of predictors and 1).

2. **Sum of Squared Residuals (ssr):**
   - Null model: 12.32 × 10¹⁵ (high residual variance indicates poor fit).
   - Unrestricted model: 3.98 × 10¹⁵ (lower residual variance shows better fit).

3. **Difference in Degrees of Freedom (df_diff):**
   - 6 (the number of predictors added in the unrestricted model).

4. **Sum of Squares Difference (ss_diff):**
   - 8.34 × 10¹⁵ (substantial reduction in residual variance when adding predictors).

5. **F-statistic (F):**
   - 5,375.66, a very large value, indicating a strong effect of the predictors on the model.

6. **P-value (Pr(>F)):**
   - 0.00 (practically less than 0.001), meaning the improvement in model fit is highly statistically significant.

**Interpretation:**
The unrestricted model, which includes predictors, explains significantly more variance in the selling price compared to the null model. The very small p-value (< 0.001) confirms that this improvement is not due to random chance. The large F-statistic (5,375.66) highlights the strong contribution of the predictors.

**Conclusion:**
The unrestricted model is a much better fit for the data, and the predictors collectively play a significant role in explaining the variance in selling price.
